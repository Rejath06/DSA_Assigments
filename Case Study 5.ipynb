# â— Check whether all the zones generate the same amount of sales.

We need to set a null hypothesis and alternate hypothesis.     
Null Hypothesis, H0 = all zones generate same amount of sales    ,   
Alternate Hypothesis, H1 = atleast one zone generate different amount of sales

```python
from scipy.stats import f_oneway
stat, p = f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

```python
stat
```

```python
p
```

```python
if(p>0.05):
    print('all zones generate same amount of sales')
else:
    print('atleast one zone generate different amount of sales')
```

From above testing we can clearly says that atleast one zone generate different amount of sales
