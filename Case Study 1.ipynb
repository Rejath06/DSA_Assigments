```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# loading the sample file

```python
car_data=pd.read_csv('cars_data.csv')
```

```python
car_data
```

# Check any null values present in the data

```python
car_data.isnull().sum()
```

There is no null values in the given data sheet

# Male vs Female buyers participated in the sales

```python
buyer=car_data['Buyer Gender'].value_counts()
print(buyer)
buyer.plot(kind='bar', figsize=(8,6))
```

There are 5052 Female buyers and 4948 Male buyers in the list

# Top 5 cars based on the sales experience

```python
top_cars=car_data.nlargest(5, ['Sale Price'])
top_cars
```

The Top 5 cars Based on the Selling price are Volkswagen Jetta 3, Audi A6, Lexus LX, Mazda MX-6 and Lexus ES

# Least Reselling Cars

```python
small_cars=car_data.nsmallest(5, ['Resell Price'])
small_cars
```

The Least reselling cars are Chevrolet 1500, GMC Savana 1500, Jaguar XF, Mistibishi 3000GT and Infiniti G37
